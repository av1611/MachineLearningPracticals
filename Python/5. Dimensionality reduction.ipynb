# Dimensionality reduction

```python
## importing the required packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

images = np.load('MNIST/mnist_images.npy')
labels = np.load('MNIST/mnist_labels.npy')
```

```python
# Setting plotting parameters
cm = plt.cm.get_cmap('RdYlBu')
plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
from sklearn.decomposition.pca import PCA
p = PCA(2)
images_projected = p.fit_transform(images)
plt.scatter(images_projected[:,0], images_projected[:,1], c=labels, s=8, cmap = cm);
```

```python
from sklearn import manifold # for t-SNE

print("Computing t-SNE embedding")
tsne = manifold.TSNE(n_components=2, n_iter = 200, random_state=0)
np.set_printoptions(suppress=True)
images_tsne = tsne.fit_transform(images)
plt.scatter(images_tsne[:,0], images_tsne[:,1], c=labels, s=8, cmap = cm);
```
